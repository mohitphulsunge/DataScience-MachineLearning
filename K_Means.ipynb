```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.cluster import KMeans
import imageio
parrot = imageio.imread("bird_small.png")
```

Importing dataset

```python
mat = loadmat("ex7data2.mat")
X = mat["X"]
X
```

Visualizing the dataset

```python
plt.scatter(X[:,0], X[:,1])
```

Apply k-means

```python
clusters = KMeans(n_clusters=3, n_init=20)
clusters.fit(X)
print(clusters.cluster_centers_)
plt.scatter(X[:,0], X[:,1])
plt.scatter(clusters.cluster_centers_[:,0], clusters.cluster_centers_[:,1],cmap = 'r')
```

Parition the data

```python
red_data = X[clusters.labels_ == 0]
plt.scatter(red_data[:,0], red_data[:,1])

green_data = X[clusters.labels_ == 1]
plt.scatter(green_data[:,0], green_data[:,1])

blue_data = X[clusters.labels_ == 2]
plt.scatter(blue_data[:,0], blue_data[:,1])
```

Image compression

```python
fig, ax = plt.subplots()
ax.imshow(parrot)
parrot = parrot.reshape(128**2, 3)
```

Plot the data in 3d graph

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
x = parrot[:,0]
y = parrot[:,1]
z = parrot[:,2]
ax.set_zlabel("Blue")
ax.set_ylabel("Green")
ax.scatter(x, y, z, s=1, color=parrot/255.0)
```

Apply K means clustering

```python
color_clust = KMeans(n_clusters=16, n_init=20)
color_clust.fit(parrot)
color_clust.cluster_centers_
```

Visualize the colors

```python
fig = plt.figure()
ax = plt.axes(projection ='3d')
x = parrot[:,0]
y = parrot[:,1]
z = parrot[:,2]
ax.set_zlabel("Blue")
ax.set_ylabel("Green")
ax.scatter(x, y, z, s=1, c=color_clust.labels_)
```

Reconstruct the image from k means centroid

```python
color_clust.cluster_centers_
newcolors = np.round(color_clust.cluster_centers_)
np_parrot = np.array(parrot)
compressed_parrot = np.apply_along_axis(map_color, 1, np_parrot)
  
```

```python
def map_color(arr):
    cluster_no = color_clust.predict(arr.reshape(1,-1))
    return newcolors[cluster_no[0]]
```

Visualize parrot

```python
fig, ax = plt.subplots()
compressed_parrot = compressed_parrot.astype(np.uint8) 
ax.imshow(compressed_parrot.reshape(128, 128, 3))
```
